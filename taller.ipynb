# Obtener la forma de una matriz

Imagínemos que tenemos una hoja de papel con una cuadrícula de números, algunos en filas y otros en columnas. Quieres saber cuántas filas y columnas hay en total. Para hacerlo, simplemente contarías cuántas filas hay y cuántas columnas hay en esa hoja de papel. Bueno, eso es exactamente lo que hace la función `obtener_forma(matriz)` en mi código.



### Como funciona obtener la forma de una matriz

Imagina que tienes una hoja de papel con una cuadrícula de números. Algunos números están en filas horizontales y otros en columnas verticales. Quieres saber cuántas filas y cuántas columnas tiene esa hoja de papel. Para hacerlo, contarías cuántas líneas horizontales (filas) y cuántas líneas verticales (columnas) hay en total. Eso es lo que hace la función `obtener_forma(matriz)`.

Ahora, vamos a ver cómo funciona el código de la función:

1. Verificación inicial:
   - La primera línea de código `if type(matriz) != list or not matriz:` verifica si la variable `matriz` no es una lista o si está vacía.
   - En nuestra analogía, esto es como comprobar si la hoja de papel no es realmente una hoja de papel o si está en blanco. En cualquiera de esos casos, no podríamos contar filas ni columnas, por lo que simplemente devolvemos una lista vacía `[]`.

2. Conteo de filas:
   - Si la matriz no es una lista vacía, entonces podemos continuar.
   - La línea `return [len(matriz), *obtener_forma(matriz[0])]` realiza dos acciones:
     - Primero, cuenta cuántas líneas horizontales (filas) hay en total en nuestra hoja de papel. Esto se hace con `len(matriz)`, que nos da el número de filas.
     - Luego, para cada fila, miramos cuántos números hay en esa fila. Esto es similar a contar cuántos números hay en cada fila de nuestra hoja de papel.
     - Finalmente, devolvemos una lista que contiene el número total de filas y los detalles de cada fila.

3. Repeticion:
   - Después de contar el número total de filas, necesitamos contar cuántas columnas hay en cada fila.
   - Esto es como mirar cada fila individualmente en nuestra hoja de papel y contar cuántos números hay en cada una.
   - Para hacer esto, tomamos la primera fila de nuestra hoja de papel y contamos sus columnas.
   - Luego, repetimos este proceso para las filas restantes.
   - Al final, obtenemos una lista que contiene el número total de filas y los detalles de cuántas columnas hay en cada fila.


````python
def obtener_forma(matriz):
    if type(matriz) != list or not matriz:
        return []
    return [len(matriz), *obtener_forma(matriz[0])]
````

## Rotar una matriz

Imagina que tenemos una hoja de papel con números escritos en ella. Ahora, quieres girar esa hoja de papel para leer los números de manera diferente. Por ejemplo, si antes leías de izquierda a derecha, ahora quieres leer de arriba abajo. Eso es lo que hace la función `rotar_matriz(matriz)` en mi código.



### Rotar una matriz

Imagina que tienes una hoja de papel con una cuadrícula de números. Ahora, quieres girar esa hoja de papel para leer los números de una manera diferente. Por ejemplo, si antes leías de izquierda a derecha, ahora quieres leer de arriba hacia abajo. Eso es lo que hace la función `rotar_matriz(matriz)`.

Veamos cómo funciona el código de la función:

1. Desempaquetar la matriz:
   - La función `zip(*matriz)` toma la matriz original y "desempaqueta" sus elementos.
   - Es como si tomaras cada fila de la hoja de papel y la extendieras en una línea recta.

2. Girar la matriz:
   - Al desempaquetar los elementos de la matriz, estamos efectivamente intercambiando las filas y las columnas.
   - Ahora, en lugar de tener las filas horizontales de la hoja de papel, tenemos las columnas verticales.
   - Esto es lo que significa "rotar" la matriz: cambiar la orientación de las filas y las columnas.

3. Reorganizar en una nueva matriz:
   - La función `list(zip(*matriz))` reorganiza estos elementos desempaquetados en una nueva matriz.
   - Ahora, la nueva matriz tiene las filas que eran columnas en la matriz original, y viceversa.
   - Esto nos da una "rotación" completa de la matriz original.



````python
def rotar_matriz(matriz):
    return list(zip(*matriz))
````

## Multiplicar dos matrices

Jajaja de nuevo, imaginemos que tenemos dos hojas de papel con números escritos en ellas, queremos encontrar el resultado de multiplicarlas juntas. Para hacerlo, hay un proceso específico que debes seguir. Eso es exactamente lo que hace la función `multiplicar_matrices(matriz1, matriz2)` en mi código.



### Multiplicar dos matrices: Prueba de escritorio

1. Preparación:
   - Tengo dos matrices: `matriz1` y `matriz2`.
   - Primero, necesito girar la segunda matriz para que sea más fácil multiplicarla con la primera.

2. Girar la segunda matriz:
   - Para girar la matriz, uso la función `rotar_matriz(matriz2)` y guardo el resultado en `matriz_rotada`.

3. Verificación de la compatibilidad:
   - Verifico si el número de columnas de `matriz1` es igual al número de filas de `matriz_rotada`.
   - Si no son compatibles, devuelvo un resultado nulo y termino.

4. Multiplicación:
   - Ahora, puedo comenzar a multiplicar las matrices.
   - Recorreré cada fila de `matriz1` y cada columna de `matriz_rotada`.
   - Para cada elemento de la matriz resultante, realizaré una serie de multiplicaciones y sumas.

5. Calculando un elemento de la matriz resultante:
   - Tomo la primera fila de `matriz1` y la primera columna de `matriz_rotada`.
   - Multiplico el primer elemento de la fila por el primer elemento de la columna, y así sucesivamente.
   - Sumo estos productos para obtener el valor del primer elemento de la matriz resultante.

6. Construcción de la matriz resultante:
   - A medida que calculo los elementos de la matriz resultante, los voy agregando a la lista `producto`.

7. Resultado:
   - Después de calcular todos los elementos, devuelvo la matriz resultante.

````python
def multiplicar_matrices(matriz1, matriz2):
    matriz_rotada = rotar_matriz(matriz2)
    forma_matriz1 = obtener_forma(matriz1)
    forma_matriz2 = obtener_forma(matriz2)
    if forma_matriz1[1] != forma_matriz2[0]:
        return None
    producto = []
    altura_matriz1 = len(matriz1)
    for fila_matriz1 in range(altura_matriz1):
        fila_producto = []
        ancho_matriz_rotada = len(matriz_rotada)
        for columna_rotada in range(ancho_matriz_rotada):
            suma = 0
            for indice_elemento in range(len(matriz1[fila_matriz1])):
                suma += matriz1[fila_matriz1][indice_elemento] * matriz_rotada[columna_rotada][indice_elemento]
            fila_producto.append(suma)
        producto.append(fila_producto)
    return producto
````
